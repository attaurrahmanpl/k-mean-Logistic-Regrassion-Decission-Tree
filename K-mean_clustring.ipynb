```python
# import Libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd 
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler
```

```python
# import data
data = pd.read_csv("https://raw.githubusercontent.com/codebasics/py/master/ML/13_kmeans/income.csv")

```

```python
data.head()
```

```python
# plot all the data on scatter plot

plt.scatter(data['Age'] , data['Income($)'])
plt.show()
```

```python
# in this case we are creating three clusters
km = KMeans(n_clusters=3)
```

```python
km
```

```python
y_predicted = km.fit_predict(data[['Age' , 'Income($)']])
y_predicted
```

```python
data['cluster'] = y_predicted
data.head()
```

```python
data1 = data[data.cluster == 0]
data2 = data[data.cluster == 1]
data3 = data[data.cluster == 2]

plt.scatter(data1.Age , data1['Income($)'] , color='red')
plt.scatter(data2.Age , data2['Income($)'] , color='black')
plt.scatter(data3.Age , data3['Income($)'] , color='green')

plt.xlabel('Age')
plt.ylabel('income($)')
plt.show()
```

```python
scaler = MinMaxScaler()
scaler.fit(data[['Income($)']])

data['Income($)'] = scaler.transform(data[['Income($)']])
# data['income($)'] = scaler.transform(data['income($)'])


# data
# scale age data
scaler.fit(data[['Age']])
data[['Age']] = scaler.transform(data[['Age']])

data
```

```python
km = KMeans(n_clusters=3)
y_predicted = km.fit_predict(data[['Age' , 'Income($)']])
y_predicted
```

```python
data['cluster'] = y_predicted
data
```

```python
data1 = data[data.cluster == 0]
data2 = data[data.cluster == 1]
data3 = data[data.cluster == 2]

plt.scatter(data1.Age , data1['Income($)'] , color='red')
plt.scatter(data2.Age , data2['Income($)'] , color='black')
plt.scatter(data3.Age , data3['Income($)'] , color='green')

plt.xlabel('Age')
plt.ylabel('income($)')
plt.show()
```
